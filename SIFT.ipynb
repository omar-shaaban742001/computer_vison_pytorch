# import libraries 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

# Get the dataset

```python
path = "F:\\programming\\computer vision nanodegree\\datasets\\images"

#load image 1
img1 = cv2.imread(path + '\\P1010517.jpg')  
img1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)

#load image 2
img2 = cv2.imread(path + '\\P1010520.jpg')  
img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
```

```python
# setup the plot size 
plt.figure(figsize=(10, 10))  

#put the first image in the first place 
plt.subplot(1 , 2 , 1)
plt.imshow(img1)

#put the second image to the next place 
plt.subplot(1 , 2 , 2)
plt.imshow(img2)

plt.show()
```

# SIFT (scale invarient features transform)

```python
#first convert the images into gray scale 
img1_gray = cv2.cvtColor(img1 , cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2 , cv2.COLOR_BGR2GRAY)
```

```python
#plot the original images with gray images 
# setup the plot size 
plt.figure(figsize=(10, 10))  

#first original image 
plt.subplot(2 , 2 , 1)
plt.imshow(img1)
plt.title("First Image")
#gray scale for the first image 
plt.subplot(2 , 2 , 2)
plt.imshow(img1_gray , cmap = 'gray')
plt.title('Gray scale for the first image')

#second original image
plt.subplot(2 , 2 , 3)
plt.imshow(img2)
plt.title("Second Image")

#gray scale for the second image 
plt.subplot(2 , 2 , 4)
plt.imshow(img2_gray , cmap='gray')
plt.title('Gray scale for the second image')

#plot the images 
plt.show()
```

```python
#get an instant of SIFT
sift = cv2.xfeatures2d.SIFT_create()

#compute the sift 
key_points_1 , descriptors_1 = sift.detectAndCompute(img1_gray, None)
key_points_2 , descriptors_2 = sift.detectAndCompute(img2_gray, None)
```

```python
len(key_points_1), len(key_points_2)
```

# Plot the images with there keypoints

```python
# setup the plot size 
plt.figure(figsize=(10, 10))  

#first original image 
plt.subplot(2 , 2 , 1)
plt.imshow(img1)
plt.title("First Image")

#gray scale for the first image 
plt.subplot(2 , 2 , 2)
img_1 = cv2.drawKeypoints(img1_gray,key_points_1, img1.copy() , flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_1)
plt.title('Key points for the first image')

#first original image 
plt.subplot(2 , 2 , 3)
plt.imshow(img2)
plt.title("Second Image")

#gray scale for the first image 
plt.subplot(2 , 2 , 4)
img_2 = cv2.drawKeypoints(img2_gray,key_points_2, img2.copy() , flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img_2)
plt.title('Key points for the second image')

#plot the images
plt.show()
```
